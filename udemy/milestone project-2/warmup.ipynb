```python
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':11, 'Queen':12, 'King':13, 'Ace':14}
```

```python
class Card:
    
    def __init__(self,suit,rank):
        self.suit=suit
        self.rank=rank
        self.value=values[rank]
        
    def __str__(self):
        
        return self.rank + ' of ' + self.suit
        
```

```python
import random
```

```python
class Deck:
    def __init__(self):
        self.all_cards=[]
        for suit in suits:
            for rank in ranks:
                c=Card(suit,rank)
                self.all_cards.append(c)
                
    def shuffle(self):
        random.shuffle(self.all_cards)
        
    def dealone(self):
        return self.all_cards.pop()
```

```python
class Player:
    
    def __init__(self,name):
        self.name=name
        self.player_cards=[]
        
    def add_cards(self,card):
        if type(card)==type([]):
            self.player_cards.extend(card)
        else:
            self.player_cards.append(card)
    
    def remove_one(self):
        return self.player_cards.pop(0)
        
    def __str__(self):
        return f"{self.name} has {len(self.player_cards)} cards"
```

# A Mini Game

```python
import time
```

```python
i=0
n=int(input("Enter the number of rounds: "))
```

```python
while i<n:
    print("Round ",i+1)
    choice1=one.remove_one()
    choice2=two.remove_one()
    if (choice1.value==choice2.value):
        print("Round {} is draw".format(i+1))
        one.player_cards.insert(0,choice1)
        two.player_cards.insert(0,choice2)  
    elif choice1.value>choice2.value:
        print("player 1 wins the round -",i+1)
        one.add_cards([choice2,choice1])
    else:
        print("player 2 wins the round -",i+1)
        two.add_cards([choice1,choice2])
    i+=1
if len(one.player_cards)>len(two.player_cards):
    print("Player One wins")
else:
    print("Player Two wins")
```
